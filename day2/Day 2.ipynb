### Chercher sur Google : "world population fusion table"

### Ensuite download du fichier

```python
df = pd.read_csv('world population.csv')
```

```python
df.sort_values('Land area', ascending=False)
```

```python
df['Population'] = df['Population'] \
    .apply(lambda x: int(x.replace(',', '')))
```

```python
df['Land area'] = df['Land area'] \
    .apply(lambda x: int(x.replace(',', '')))
df['Density'] = df['Density'] \
    .apply(lambda x: int(x.replace(',', '')))
```

```python
df.sort_values('Land area', ascending=False)
```

### Create a new density column and compute the difference with the given density

```python
df['new_density'] = df['Population'] / df['Land area']
```

```python
df['diff'] = abs((df['Density'] - df['new_density']) / df['Density'])
```

```python
df.sort_values('diff', ascending=False)
```
